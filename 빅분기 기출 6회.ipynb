1.   주어진 데이터를 이용하여 이동시간(출동시간과 도착시간의 차이)이 평균적으로 가장 오래 걸린 구급대의 이동시간을 구하시오




```python
import pandas as pd
import numpy as np

df = pd.read_csv('https://raw.githubusercontent.com/YoungjinBD/dataset/main/data_6_1_1.csv')
print(df.head())
print()
print(df.info())
print()
print(df.describe())
```

```python
# 데이터의 전처리 - 수치형으로 변환
df['출동시간'] = pd.to_datetime(df['출동시간'])
df['도착시간'] = pd.to_datetime(df['도착시간'])
```

```python
df['이동시간'] = df['도착시간'] - df['출동시간']
df['이동시간'] = df['이동시간'].dt.total_seconds()/60
print(df)
```

```python
#grouped = df.groupby('구급대')#['이동시간'].mean()
#grouped.size()

grouped = df.groupby('구급대')['이동시간'].mean()
grouped
```

```python
print(round(grouped.max()))
```

3. 주어진 데이터를 이용하여 연도별 총 범죄 건수의 월평균 값을 구하고, 그 값이 가장 큰 연도의 총 범죄 건수의 월평균 값을 집계하시오(반올림하여 정수로 출력)

```python
df = pd.read_csv('https://raw.githubusercontent.com/YoungjinBD/dataset/main/data_6_1_3.csv')
print(df.head(15))
print()
print(df.info())
```

```python
df['연도'] = df['날짜'].str[:4].astype(int)
df.head()
```

```python
# Label과 인덱스의 활용을 통한 .loc
# 행번호(row number)로 선택하는 방법(.iloc)

df['총범죄수']= df.iloc[:,1:-2].sum(axis=1)
```

```python
group = df.groupby('연도')['총범죄수'].mean()
group
```

```python
print(round(group.max()))
```

**제2유형**

학습 530명에 대한 학습용(train)테이터를 이용하여 학생 구분(유치원생,초등학생,중학생,고등학생,대학생) 예측 모형을 만든 후, 이를 평가용(test) 데이터에 적용하여 얻은 230명 학생의 구분을 다음과 같은 형식의 CSV 파일로 생성하시오.
(제출한 모델의 성능은 f1-marco)평가에 따라 채점


```python
import pandas as pd

train = pd.read_csv('https://raw.githubusercontent.com/YoungjinBD/dataset/main/data_6_2_train.csv')
test = pd.read_csv('https://raw.githubusercontent.com/YoungjinBD/dataset/main/data_6_2_test.csv')

print(train.info())
print()
print(train.head())
print()
print(train.describe())

col_scale =['일평균수면시간','일평균학습시간','일평균스마트폰사용시간','몸무게','기초대사량']
col_encoder =['키','체격','거주지역']
```

```python
# 결측치 처리, 수치데이터 스케일링, 텍스트 인코딩(키,거주지역)

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
train['일평균수면시간'] = scaler.fit_transform(train[['일평균수면시간']])
test['일평균수면시간'] = scaler.fit_transform(test[['일평균수면시간']])

train['일평균학습시간'] = scaler.fit_transform(train[['일평균학습시간']])
test['일평균학습시간'] = scaler.fit_transform(test[['일평균학습시간']])

train['일평균스마트폰사용시간'] = scaler.fit_transform(train[['일평균스마트폰사용시간']])
test['일평균스마트폰사용시간'] = scaler.fit_transform(test[['일평균스마트폰사용시간']])

train['몸무게'] = scaler.fit_transform(train[['몸무게']])
test['몸무게'] = scaler.fit_transform(test[['몸무게']])

train['기초대사량'] = scaler.fit_transform(train[['기초대사량']])
test['기초대사량'] = scaler.fit_transform(test[['기초대사량']])

```

```python
temp=train['구분']
train = pd.get_dummies(train.drop(['구분'],axis=1))
train['구분'] = temp

test = pd.get_dummies(test)
test.info()



#from sklearn.preprocessing import LabelEncoder

#encoder = LabelEncoder()
#train['키'] = encoder.fit_transform(train[['키']])
#test['키'] = encoder.transform(test[['키']])

#train['체격'] = encoder.fit_transform(train[['체격']])
#test['체격'] = encoder.transform(test[['체격']])

#train['거주지역'] = encoder.fit_transform(train[['거주지역']])
#test['거주지역'] = encoder.transform(test[['거주지역']])
```

```python
from sklearn.preprocessing import StandardScaler

for col in col_scale:
  scaler = StandardScaler()
  scaler.fit(train[[col]])

  train[col] = scaler.transform(train[[col]])
  test[col] = scaler.transform(test[[col]])


from sklearn.preprocessing import LabelEncoder

for col in col_encoder :
  encoder = LabelEncoder()
  encoder.fit(train[col])
  train[col] = encoder.transform(train[[col]])
  test[col] = encoder.transform(test[[col]])

print(train.head())
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(train,train['구분'],test_size=0.2,stratify = train['구분'])
x_train.drop(['구분'],axis=1, inplace=True)
x_test.drop(['구분'],axis=1, inplace = True)
print(x_train.head())
print()
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(random_state=123)
rf.fit(x_train,y_train)

pred = rf.predict(x_test)

from sklearn.metrics import f1_score

f1_score = f1_score(y_test,pred,average='macro')
print(f1_score)
```

```python
pred_test = rf.predict(test)
result = pd.DataFrame({'구분예측':pred_test})
print(result)

#result.to_csv('reuslt.csv',index = False)
```
