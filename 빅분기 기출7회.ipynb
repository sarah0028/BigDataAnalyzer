```python
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/BigData/07.01.03-apartment_prices_dataset.csv')
print(df.head())
print()
print(df.info())
print()
print(df.describe())
```

```python
#print(df.columns[:-1])
#이상치는 (1Q-1.5IQR) 보다 작거나 (3Q+1.5IQR)보다 큰 값이며, IQR은 Q3에서 Q1을 뺀 값임)

oct_reult = {}
#oct =[]
for col in df.columns[:-1]:
  Q1 = df[col].quantile(.25)
  Q3 = df[col].quantile(.75)
  IQR = Q3-Q1
  #oct.append((df[col]< (Q1-1.5*IQR)) |( df[col]>(Q3+1.5*IQR)).sum())
  oct = ((df[col]< (Q1-1.5*IQR)) | (df[col]>(Q3+1.5 * IQR))).sum()
  oct_reult[col] = oct


print(oct_reult)
```

```python
print( max(oct_reult,key= oct_reult.get) )
```

```python
import pandas as pd

train = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/BigData/07.02.01-sales_train_dataset.csv')
test = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/BigData/07.02.02-sales_test_dataset_x.csv')
test_y = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/BigData/07.02.03-sales_test_dataset_y.csv')
```

```python
# 결측치, 수치형 스케일링, 인코딩 - randomforestregresstion

train = train.drop(['BranchName'],axis=1)
col_num = ['Population', 'IncomeGeneratingPopRatio','AverageIncome']
col_cat = ['City','IndustryType']

train.info()
```

# 제3유형 1번
키(Height), 몸무게(Weight), 발사이즈(ShoeSize)를 이용하여 남녀(Sex) 성별을 분류하는 로지스틱 회귀모형을 만들고자 한다. 주어진 데이터세트를 이용하여 로지스틱 회귀모형을 만들고 아래의 질문에 답하시오. 단, 주어진 데이터세트의 2/3를 훈련용으로, 1/3을 테스트용으로 분할하여 작업하시오.
1.  훈련용 데이터를 이용하여 'Height', 'Weight', 'ShoeSize'를 설명변수로 사용하는 로지스틱 회귀 모형을 만들고 Weight의 오즈비를 계산하시오.
2.  모델의 로짓우도(log-likelihood)를 구하시오.
3. 위 (2)에서 만든 모델로 평가용 데이터를 예측한 결과와 실제값의 오차율을 구하시오. (소수 네자리에서 반올림)

```python
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/BigData/07.03.01-gender_prediction_dataset.csv')
df.head()
```

```python
df.info()
```

```python
# 훈련용 2/3 : 200개, 테스트용: 1/3 100개

from sklearn.preprocessing import LabelEncoder

train = df.iloc[:200,:]
test = df.iloc[200:,:]

encoder = LabelEncoder()
train['Gender'] = encoder.fit_transform(train[['Gender']])
test['Gender'] = encoder.transform(test[['Gender']])

print(train.shape, test.shape)
print(train.head())
```

```python
#훈련용 데이터를 이용하여 'Height', 'Weight', 'ShoeSize'를 설명변수로 사용하는 로지스틱 회귀 모형을 만들고 Weight의 오즈비를 계산하시오.

from statsmodels.formula.api import logit

result = logit('Gender ~ Height + Weight + ShoeSize',data = train).fit()
```

```python
import numpy as np
print(result.summary())
print(np.exp(result.params['Weight']))
```

```python
print('로짓우드 : ', result.llf, '잔차이탈도 = -2*llf : ', -2*result.llf)
```

```python
proba = result.predict(test)
y_answer = test['Gender']

y_pred = (proba >= 0.5).astype(int)
result2 = (y_pred != y_answer).mean()   # 오차율

print(proba)
```
