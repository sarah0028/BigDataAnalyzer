**독립성 검정 - 카이제곱 검정**

* 범주형 데이터에 대한 통계적 가설 검정을 수행
* 두 개 이상의 범주형 변수 간의 관계를 분석하거나 관찰된 빈도와 기대 빈도 사이의 차이를 확인하는데 활용

**stats.chi2_contingency()**
1. from scipy.stats import chi2_contingency
2.  범주형 데이터를 pd.crosstab(df.col1, df.col2) 교차표를 만듦
3. chi, p_val, df, expected 순으로 변수를 받음

```python
# 타이타닉 데이터셋을 이용하여 다음 아래 질문에 답하시오
# (1) 생존 여부와 성별 간의 독립성 검정
# (2) 생존 여부와 연령대 간의 독립성 검정

import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/YoungjinBD/dataset/main/titanic.csv')

df.info()
```

```python
table = pd.crosstab(df['Sex'],df['Survived'])
```

```python
from scipy.stats import chi2_contingency

chi, p_val, df, exp = chi2_contingency(table)
print(chi,p_val,df,exp)

# p_val가 매우 작으므로 귀무가설 기각 => 연관성이 있다

```

# 대응표본의 평균 차이 검정(t- 검정)


*   두 집단의 평균 차이를 비교하고, 이 차이가 통계적으로 유의미한지 확인


* from scipy import stats

* 귀무가설 : 차이가 없다
* 대립가설 : 차이가 있디


```python
from scipy import stats

df1 = [7,3,4,5,2,1,6,6,5,4]
df2 = [8,4,5,6,2,3,6,8,6,5]

#diff = df1-df2

t_stat, p_value = stats.ttest_rel(df1,df2)
print(t_stat,p_value)

# 귀무가설 기각 :
```

# 정규성 검정

```python
from scipy import stats

df = [7,3,4,5,2,1,6,6,5,4]

p_value = stats.shapiro(df)
print(p_value)

# 귀무가설 : 정규분포를 따른다
# 대립가설 : 정규분포를 따르지 않는다
```
